```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('Influencer.xlsx')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Followers'].unique()
```

```python
df['Avg. Likes'].unique()
```

```python
df['Avg. Likes']=[float(i[:-1])*1000000 if i[-1]=='m' else float(i[:-1])*1000 for i in df['Avg. Likes']]
```

```python
df['Posts']=[str(i) if type(i)!=str else i for i in df['Posts']]
```

```python
df['Followers']=[float(i[:-1])*1000000 for i in df['Followers'] if i[-1]=='m']
```

```python
df['Posts']=[float(i[:-1])*1000 if i[-1]=='k' else float(i) for i in df['Posts']]
```

```python
df['New Post Avg. Likes']=[str(i)  if type(i)!=str else i for i in df['New Post Avg. Likes']]
```

```python
df['New Post Avg. Likes']=[float(i[:-1])*1000000 if i[-1]=='m' else float(i[:-1])*1000 if i[-1]=='k' else float(i) for i in df['New Post Avg. Likes']]
```

```python
df['Total Likes']=[float(i[:-1])*1000000 if i[-1]=='m' else float(i[:-1])*1000000000 for i in df['Total Likes']]
```

```python
df['Channel Info']=df['Channel Info'].apply(str)
import re
def clean_name(r):
    r=re.sub('\n','',r)
    return r
```

```python
df['Channel Info']=df['Channel Info'].apply(lambda x:clean_name(x))
```

# 4. Who are the top 10 influencers in the given dataset based on the following features

```python
foll=df.nlargest(10,'Followers')
sns.barplot(x=foll['Channel Info'], y=foll['Followers'])
plt.xlabel('Channel Info')
plt.xticks(rotation=90)
plt.ylabel("Followers")
plt.title("Top 10 Influencers based on Followers")
plt.show()
```

# 5. Describe the relationship between the following pairs of features using a suitable graph

#  Followers and Total Likes

```python
sns.lmplot(x='Followers',y='Total Likes', data=df)
```

```python
sns.lineplot(x='Followers',y='Total Likes', data=df)
```

```python
sns.boxplot(df['Followers'])
```

```python
l1=df
Q1 = l1['Followers'].quantile(0.25)
Q3 = l1['Followers'].quantile(0.75)
IQR = Q3 - Q1
ll=Q1-1.5*IQR
ul=Q3+1.5*IQR
l=l1[(l1['Followers']>=ll) & (l1['Followers']<=ul)]
```

```python
plt.scatter(l['Followers'],l['Total Likes'])
```

Followers has many outliers. So after clearing the outliers it is possible to see that 'Followers' and 'Total Likes' are highly correlated. So if the Followers are high the the Total Likes is also high.
